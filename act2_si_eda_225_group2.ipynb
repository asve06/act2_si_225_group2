### 3. Drop columns that are not relevant

We decided to drop the columns **review_text_liked, review_text_disliked, full_review, and en_full_review**, as they either repeat information already present in **review_title** or merge data redundantly, adding no unique value. We also removed **found_helpful** and **found_unhelpful**, since they are always 0 and don't provide meaningful insight for analyzing guest satisfaction or trends. Additionally, **owner_resp_text** and **original_lang** were dropped as they are not relevant to understanding the guest experience. We kept **username, room_view, stay_duration, stay_type, review_post_date, review_title, rating**, and **user_country** to retain key contextual and demographic information that can help analyze trends and satisfaction patterns effectively.

```python
columns_to_drop = ['review_text_liked', 'review_text_disliked', 'full_review', 'en_full_review', 'found_helpful', 'found_unhelpful', 'owner_resp_text', 'original_lang']
df.drop(columns = columns_to_drop, inplace= True)
```

### 5. Splitting of Review Date into Month and Year

```python
df['review_post_date'] = pd.to_datetime(df['review_post_date'])
df.info()
```

```python
df['month'] = df['review_post_date'].dt.month
df['year'] = df['review_post_date'].dt.year
```

#Plots

```python
import matplotlib.pyplot as plt
```

### Distribution of Stay Types - Ashley Vega

```python
stay_counts = df['stay_type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    stay_counts,
    labels=stay_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Distribution of Stay Types")
plt.show()
```

This pie chart distribution of the type of hotel guests suggest that the hotel Kaudalini primarily attracts couples seeking leisure stays, followed by the family & group visits are less frequent. Understanding this trend can help management tailor services and marketing strategies to satisfy the dominant guest segments.

### Average rating per year - Leynder Sanchez

```python
avg_rating_year = df.groupby('year')['rating'].mean()

plt.plot(avg_rating_year.index, avg_rating_year.values, marker='o', color='orange')
plt.title('Trend of Average Rating per Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.ylim(0,10)
plt.grid(True)
plt.show()
```

This line plot demonstrates that the hotel has maintained high ratings over the past three years, consistently ranging between 9 and 10, which indicates guest satisfaction with the hotel's quality and services.

### Reviews per month - Jimmy Ochoa

```python
reviews_per_month = df.groupby('month').size()
reviews_per_month.plot(kind='bar', color='skyblue')
plt.title("Number of Reviews per Month")
plt.xlabel("Month")
plt.ylabel("Number of Reviews")
plt.show()
```

This bar chart shows the number of reviews each month. **October** has the most reviews, while **June** and **November** have the least. From **February** to **April**, the number of reviews stays quite high.
